Analyze the text reviews to identify the most
 common positive and negative keyword


```python
positive_Data = df[df['Aggregate rating']>2.5]
print(positive_Data)
```

```python
positive_key = positive_Data['Rating text'].unique()
print(positive_key)
```

```python
plt.hist(positive_Data['Rating text'],label=positive_key)
plt.show()
```

```python
neg_data = df.loc[(df['Aggregate rating']<2.5) & (df['Aggregate rating']>0)]
neg_key = neg_data['Rating text'].unique()
print(neg_key)
```

```python
plt.hist(neg_data['Rating text'],label = neg_key)
plt.show()
```
