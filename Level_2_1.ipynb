 Analyze the distribution of aggregate
 ratings and determine the most common
 rating range.

```python
aggregate_c = df['Aggregate rating'].value_counts()
aggregate_c
```

```python
a = aggregate_c.idxmax()
print("most common range = ", a)
```

```python
plt.hist(df['Aggregate rating'],bins =30 ,color='red' ,edgecolor='black')
plt.xlabel("Aggregate rating ")
plt.ylabel("count")
plt.title("Dist of agreation rating ")
plt.show()
```
