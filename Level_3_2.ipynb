```python
import pandas as pd
import matplotlib.pyplot as  plt
```

```python
df = pd.read_csv("Dataset.csv")
df
```

dentify the restaurants with the highest and
 lowest number of votes.

```python
high_v =df['Votes'].max()
high_v
```

```python
low_v= df['Votes'].min()
low_v
```

```python
high_votes_res = df.groupby('Restaurant Name')['Votes'].sum().sort_values(ascending = False)
high_votes_res = high_votes_res.idxmax()
print("high_votes_res :",high_votes_res)
```

```python
low_votes_res= df.groupby('Restaurant Name')['Votes'].sum().sort_values()
low_votes_res =low_votes_res.idxmin()
print("Low voting  res is :",low_votes_res)
```

Analyze if there is a correlation between the
 number of votes and the rating of a
 restaurant.

```python
import numpy as np
import seaborn as sns
```

```python
pip install seaborn
```

```python
corelation = df['Votes'].corr(df['Aggregate rating'])
print(round(corelation),2)
```

```python
correl = np.corrcoef(df['Votes'],df['Aggregate rating'])
correl=np.round(correl,2)
print(correl)
```

```python
plt.figure(figsize=(10,5),dpi = 100)
sns.heatmap(correl, annot=True, fmt=".2f", linewidths=.4)

plt.show()
```

```python
sns.scatterplot(data=df,y='Votes',hue='Aggregate rating',x='Aggregate rating')
sns.regplot(data=df ,x='Aggregate rating' ,y ='Votes',scatter =False)
plt.title("cor bt aggreation and voting :0.3")
plt.show()
```
