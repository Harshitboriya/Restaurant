```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df =pd.read_csv('Dataset.csv')
df
```

```python
df.info()
```

```python
res_chain  = df.groupby('Restaurant Name').size().reset_index(name='OutletCount')
new = res_chain[res_chain['OutletCount']>1]
res_chain = new.sort_values(by='OutletCount',ascending =False)
```

```python
print(res_chain[['Restaurant Name','OutletCount']].head(10))
```

```python
plt.bar(res_chain['Restaurant Name'][:5],res_chain['OutletCount'][:5] )
plt.xlabel("Rest chains..")
plt.ylabel("Number of Outlets")
plt.title("top 5 rest chain by number")
plt.show();
```

Analyze the ratings and popularity of
 different restaurant chains.


```python
rating = df.groupby("Restaurant Name")['Aggregate rating'].mean().reset_index(name='Average Rating').sort_values(by='Average Rating',ascending =False)
rating
```

```python
votes = df.groupby("Restaurant Name")['Votes'].sum().reset_index(name='Total votes').sort_values(by='Total votes',ascending =False)
votes
```

```python
plt.bar(rating['Restaurant Name'][:5],rating['Average Rating'][:5] )
plt.xlabel("Rest chains..")
plt.ylabel("Number of Outlets")
plt.title("top 5 rest avg rating ")
plt.show();
```

```python
plt.bar(votes['Restaurant Name'][:5],votes['Total votes'][:5] )
plt.xlabel("Rest chains..")
plt.ylabel("Number of Outlets")
plt.title("top 5 rest avg rating ")
plt.show();
```
