Create a histogram or bar chart to

Calculate the percentage of restaurants
 in each price range category.



```python
price_r_c = df['Price range'].value_counts()
totle_r = len(df)
per = round((price_r_c/totle_r)*100,2)
perc_m = pd.merge(price_r_c,per,on="Price range")
perc_m
```
