```python
rating_map = {
    "Excellent" : 5,
    "Very Good" :4,
    "Good":3,
    "Average":2,
    "Poor" :1
}

df['rating numeric'] = df['Rating text'].map(rating_map)

avg_ol  = round(df[df['Has Online delivery']=='Yes']['rating numeric'].mean(),2)
print(avg_ol)

avg_not_ol = round(df[df['Has Online delivery']=='No']['rating numeric'].mean(),2)
print(avg_not_ol)

```

 Compare the average ratings of restaurants
 with and without online delivery

```python
avg_online  = round(df[df['Has Online delivery']=='Yes']['Aggregate rating'].mean(),2)
print(avg_online)

avg_not_online = round(df[df['Has Online delivery']=='No']['Aggregate rating'].mean(),2)
print(avg_not_online)

```

```python
labels =['Yes','No']
avg_rating = [avg_online,avg_not_online]
colors= 'green','red'
plt.barh(labels,avg_rating,color= colors)
plt.xlabel("avg rating")
plt.ylabel("online del")
plt.title("comparis of avg rating oaa online del")
plt.show()
```
